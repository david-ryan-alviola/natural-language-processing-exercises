4. Make function for all entries

```python
def extract_person_information(person):
    name = person.select(".name")[0]
    quote = person.select(".quote")[0]
    email = person.select(".email")[0]
    phone = person.select(".phone")[0]
    address = person.find_all("p")[3]
    
    return {'name' : name.text.strip(), 'quote' : quote.text.strip(), 'email' : email.text.strip(), 'phone' : phone.text.strip(),\
            'address' : address.text.strip()}

def get_information_by_person(soup):
    people_container = soup.select("#people")[0]
    people = people_container.select(".person")
    
    return pd.DataFrame([extract_person_information(person) for person in people])
```

```python
def extract_article_info(card, category):
    title = card.find("span", attrs={'itemprop' : "headline"})
    content = card.find("div", attrs={'itemprop' : "articleBody"})
    author = card.select(".author")[0]
    time = card.select(".time")[0]
    
    return {'title' : title.text, 'category' : category, 'content' : content.text, 'author' : author.text,\
            'timestamp' : time['content']}
```

```python
def basic_clean(string):
    cleaned_string = string
    
    cleaned_string = cleaned_string.lower()
    cleaned_string = unicodedata.normalize("NFKD", cleaned_string).encode("ascii", "ignore").decode("utf-8", "ignore")
    cleaned_string = re.sub(r"[^a-z0-9'\s]", "", cleaned_string)
    
    return cleaned_string
```

```python
original = "A $ a day Keeps the cőllectór away"
assert(basic_clean(original) == "a  a day keeps the collector away")
```

2. Define a function named tokenize. It should take in a string and tokenize all the words in the string.

```python
def tokenize(string):
    tokenized_string = string
    tokenizer = nltk.tokenize.ToktokTokenizer()
    
    return tokenizer.tokenize(tokenized_string, return_str=True)
```

3. Define a function named stem. It should accept some text and return the text after applying stemming to all the words.

```python
def stem(string):
    ps = nltk.porter.PorterStemmer()
    
    stems = [ps.stem(word) for word in string.split()]
    
    return " ".join(stems)
```

```python
original = "this is an unstemmed string"

assert(stem(original) == "thi is an unstem string")
```

4. Define a function named lemmatize. It should accept some text and return the text after applying lemmatization to each word.

```python
def lemmatize(string):
    wnl = nltk.stem.WordNetLemmatizer()
    
    lemmas = [wnl.lemmatize(word) for word in string.split()]
    
    return " ".join(lemmas)
```

5. Define a function named remove_stopwords. It should accept some text and return the text after removing all the stopwords.
    * This function should define two optional parameters, extra_words and exclude_words. These parameters should define any additional stop words to include, and any words that we don't want to remove.

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(string, extra_words=None, exclude_words=None):
    stopword_list = stopwords.words("english")
    extras = []
    excludes = []
    
    if extra_words != None:
        extras = extra_words.copy()
        
    if exclude_words != None:
        excludes = exclude_words.copy()
    
    stopword_list = stopword_list + extras
        
    for ex_word in excludes:
        if ex_word in stopword_list:
            stopword_list.remove(ex_word)

    filtered_words = [word for word in string.split() if word not in stopword_list]

    
    return " ".join(filtered_words)
```

```python
original = "biology is a complicate subject"

assert(remove_stopwords(original, extra_words=["subject"]) == "biology complicate")
assert(remove_stopwords(original, exclude_words=["subject"]) == "biology complicate subject")
assert(remove_stopwords(original, extra_words=["subject"], exclude_words=["is"]) == "biology is complicate")
assert(remove_stopwords(original) == "biology complicate subject")
```

```python
codeup_df.title = codeup_df.title.apply(remove_stopwords)
codeup_df.content = codeup_df.content.apply(remove_stopwords)
codeup_df
```

```python
for column in news_df.columns:
    if column != "timestamp":
        news_df[column] = news_df[column].apply(remove_stopwords)
        
news_df
```
