```python
import env
import pandas as pd
import requests as req
import bs4
```

```python
response = req.get("https://web-scraping-demo.zgulde.net/people")
html = response.text

soup = bs4.BeautifulSoup(html)
```

1. Get the people container

```python
people_container = soup.select("#people")[0]
```

```python
people_container
```

2. Get the individual entries within the people_container

```python
people = people_container.select(".person")
```

```python
len(people)
```

* This length matches the number of person entries on the page

3. Process the individual entries for each person

```python
person = people[0]
person
```

```python
name = person.select(".name")[0]
print(name.text)

quote = person.select(".quote")[0]
print(quote.text)

email = person.select(".email")[0]
print(email.text)

phone = person.select(".phone")[0]
print(phone.text)

address = person.find_all("p")[3]
print(address.text)
```

4. Make function for all entries

```python
def extract_person_information(person):
    name = person.select(".name")[0]
    quote = person.select(".quote")[0]
    email = person.select(".email")[0]
    phone = person.select(".phone")[0]
    address = person.find_all("p")[3]
    
    return {'name' : name.text, 'quote' : quote.text, 'email' : email.text, 'phone' : phone.text, 'address' : address.text}

def get_information_by_person(soup):
    people_container = soup.select("#people")[0]
    people = people_container.select(".person")
    
    return pd.DataFrame([extract_person_information(person) for person in people])
```

```python
get_information_by_person(soup)
```
